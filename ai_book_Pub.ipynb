```python
# âœ… 1. Install required packages if not already done
!pip install gTTS google-generativeai sentence-transformers beautifulsoup4

# âœ… 2. Imports
import os
from bs4 import BeautifulSoup
import requests
from gtts import gTTS
from IPython.display import Audio
import google.generativeai as genai

# âœ… 3. Set your Gemini API Key
os.environ['GEMINI_API_KEY'] = "AIzaSyAjuRwW8XKIGV8u2xmeQ7lPcABRuHlSSYE"
genai.configure(api_key=os.environ['GEMINI_API_KEY'])

# âœ… 4. Helper Functions
def fetch_chapter(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    paragraphs = soup.select("div#mw-content-text p")
    return "\n\n".join([p.get_text() for p in paragraphs if len(p.get_text()) > 50])

def ai_writer(text):
    model = genai.GenerativeModel("models/gemini-1.5-flash")
    prompt = f"Paraphrase and creatively rewrite the following chapter text:\n\n{text[:2000]}"
    response = model.generate_content(prompt)
    return response.text.strip()

def speak(text, filename="tts_output.mp3"):
    tts = gTTS(text)
    tts.save(filename)
    return filename
```

```python
chapter_url = "https://en.wikisource.org/wiki/The_Gates_of_Morning/Book_1/Chapter_1"
original = fetch_chapter(chapter_url)
spun = ai_writer(original)

print("âœ… Original (excerpt):", original[:400], "\n")
print("ğŸŒ€ Spun Version (excerpt):", spun[:400])
```

```python
mp3_path = speak(spun[:300], "tts_output.mp3")
Audio(mp3_path)  # ğŸ‘ˆ This will play inside Colab
```

```python
import nest_asyncio
import asyncio
from playwright.async_api import async_playwright

nest_asyncio.apply()

async def take_screenshot(url, save_path="screenshot.png"):
    async with async_playwright() as p:
        browser = await p.chromium.launch()
        page = await browser.new_page()
        await page.goto(url)
        await page.screenshot(path=save_path, full_page=True)
        await browser.close()

await take_screenshot("https://en.wikisource.org/wiki/The_Gates_of_Morning/Book_1/Chapter_1")
```
